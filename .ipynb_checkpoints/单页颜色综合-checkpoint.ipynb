# Panel rects识别

```python
# 辅助函数
def get_top_contours(contours, hierarchy):
    """
    获得顶层轮廓的下标及轮廓信息
    :param contours: 所有的轮廓信息列表
    :param hierarchy: 对应的轮廓等级信息列表，尺寸为(N,4)
    :return top_index, top_contours: 所有父轮廓的下标，轮廓信息
    """    
    top_contour_index = -1
 
    top_contours = []   # 最外层轮廓坐标
    top_index = []  # 最外层轮廓索引
    for i, contour in enumerate(contours):     # 顶层轮廓没有父轮廓
        if hierarchy[i][3] == top_contour_index:
            top_contours.append(contour)
            top_index.append(i)
    
    return top_index, top_contours


def filterContoursBySize(contours, img_h, img_w, mode=1, ratio=1/10):
    """
    根据轮廓的尺寸筛选符合要求的轮廓
    :param contours: 所有轮廓的坐标信息
    :param img_h, img_w: 原始BGR图像的高度和宽度
    :param mode: _int_ 默认使用长宽or占比方案进行筛选.若值为2，则使用&占比方案;若值为3，则使用面积占比进行筛选
    :param ratio: 设置占比比例
    :return contours, rects: 筛选后的contours及其对应的最小外界矩形坐标
    """
    total_area = img_h * img_w
    filter_contours = []
    rects = [] # 存储最小外接矩形

    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        rect = np.array([x, y, w, h])
        if mode == 1:
            if (w <= img_w*ratio) or (h <= img_h*ratio):
                pass
            else:
                rects.append(rect)
                filter_contours.append(contour)
        elif mode == 2:
            if (w <= img_w*ratio) & (h <= img_h*ratio):
                pass
            else:
                rects.append(rect)
                filter_contours.append(contour)
        elif mode == 3:
            contour_area = cv2.contourArea(contour)
            area_ratio = contour_area / total_area
            if area_ratio > ratio:
                rects.append(rect)
                filter_contours.append(contour)
            else: 
                pass
    
    return filter_contours, rects
```

```python
def correction(lst, error=10):
    i = 0
    n = len(lst)
    
    while i < n:
        # 检查当前元素与下一个元素的差值
        if i + 1 < n and abs(lst[i][1] - lst[i + 1][1]) <= error:
            # 如果差值小于等于error，则合并这两个元素
            element = min(lst[i][1], lst[i + 1][1])  # 取最小值
            lst[i][1] = element
            lst[i + 1][1] = element
        i += 1  # 跳过已经修正的元素   
    
    return lst
```

```python
def panels(image, read_mode=2, filter_mode=2, ratio=0.1):
    """
    读取image并返回panels的坐标集

    Args:
        image (_type_): 编码图像
        read_mode(_int_): 默认阅读顺序为从左到右。若值为2，则阅读顺序为从右到左
        filter_mode(_int_): 默认使用长宽&占比方案进行筛选.若值为1,则使用or占比方案;若值为3，则使用面积占比进行筛选

    Returns:
        panel_contours: panel的轮廓信息列表 
        rects: panel最小外接矩形坐标列表
    """
    
    img = np.copy(image)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    threshold = 200

    # 应用阈值处理得到二值图像
    # 检测对象为白色，背景为黑色
    _, binary = cv2.threshold(gray, threshold, 255, cv2.THRESH_BINARY_INV)

    img_h,img_w = gray.shape
    total_area = img_h * img_w

    # 查找所有轮廓，并建立等级关系
    contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    hierarchy = hierarchy.reshape(-1, 4) # reshape hierarchy

    # 查找所有的最外层轮廓
    top_index, top_contours = get_top_contours(contours, hierarchy)
    
    # 根据长宽占比筛选符合要求的panel
    panel_contours, rects = filterContoursBySize(top_contours, img_h, img_w, filter_mode, ratio)

    # 根据阅读顺序重新排序panels列表
    tran_panel_contours = []
    tran_rects = []

    sorted_index_rects = sorted(enumerate(rects), key=lambda x: x[1][1]) # 按照y从小到大
    for index, value in sorted_index_rects:
        tran_panel_contours.append(panel_contours[index])
        tran_rects.append(value)

    if(read_mode==1):
        sorted_index_rects = sorted(enumerate(correction(tran_rects)), key=lambda x: (x[1][1], x[1][0])) # 按照y,x从小到大排序
        sorted_panel_contours = []
        sorted_rects = []
        for index, value in sorted_index_rects:
            sorted_panel_contours.append(tran_panel_contours[index])
            sorted_rects.append(value)
    else:
        sorted_index_rects = sorted(enumerate(correction(tran_rects)), key=lambda x: (x[1][1], -x[1][0])) # 按照y从小到大, x从大到小
        sorted_panel_contours = []
        sorted_rects = []
        for index, value in sorted_index_rects:
            sorted_panel_contours.append(tran_panel_contours[index])
            sorted_rects.append(value)
                            
    return sorted_panel_contours, sorted_rects
```

# 简单hsv均值计算及rgb色块比例计算函数

```python
# color
# 颜色空间相关函数

import cv2
import numpy as np

def is_gray(img):
    """
    Determine whether the image is a grayscale image
    :param img: image
    :return: 1(True)/0(False)
    """
    b, g, r=cv2.split(img)
    if np.array_equal(b, g) & np.array_equal(b, r):
        return 1
    else:
        return 0


def compute_hsv_mean(hsv_image):
    """
    compute the hsv_mean of the hsv_image
    :param hsv_img: hsv_image
    :return: h_mean, s_mean, v_mean
    """
    h = hsv_image[ : , : , 0 ]
    s = hsv_image[ : , : , 1 ]
    v = hsv_image[ : , : , 2 ]
    
    # 计算H、S、V通道的均值
    mean_h = round(cv2.mean(h)[0], 2) # np.mean(h)
    mean_s = round(cv2.mean(s)[0], 2)
    mean_v = round(cv2.mean(v)[0], 2) 
    
    
    return mean_h, mean_s, mean_v
       

def hsv_mean(image, rects=None):
    """
    the hsv_mean of the image/panel
    :param img: image
    :param panels: rects of panels
    :return: h_mean, s_mean, v_mean
    """
    # 转换颜色空间从BGR到HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) 
    
    if rects is None:
        mean_h, mean_s, mean_v = compute_hsv_mean(hsv_image)
        return mean_h, mean_s, mean_v
    else:
        mean_hs = []
        mean_ss = []
        mean_vs = []
        for i in range(len(rects)):
            x, y, w, h = rects[i]
            hsv_panel = hsv_image[y:y+h, x:x+w]
            mean_h, mean_s, mean_v = compute_hsv_mean(hsv_panel)
            mean_hs.append(mean_h)
            mean_ss.append(mean_s)
            mean_vs.append(mean_v)
        return mean_hs, mean_ss, mean_vs


def compute_color_ratio(hsv_image):
    # 定义颜色的HSV范围
    color_ranges = {
        'red':((0, 43, 46), (10, 255, 255)),
        'yellow': ((26, 43, 46), (34, 255, 255)),
        'blue': ((100, 43, 46), (124, 255, 255))
    }
    
    color_ratios = {}   
    
    for color_name, (lower, upper) in color_ranges.items():
        # 创建颜色掩码
        mask = cv2.inRange(hsv_image, np.array(lower), np.array(upper))
        # kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
        # mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

        # 统计0和255出现的次数
        unique_value, counts = np.unique(mask, return_counts=True)
        if len(counts) == 2:
            true_pixels = counts[1]
        else:
            true_pixels = 0
        total_pixels = mask.size
        true_ratio = round(true_pixels/total_pixels, 3)
        
        color_ratios[color_name] = true_ratio
    
    red_color_ratio = color_ratios["red"]
    yellow_color_ratio = color_ratios["yellow"]
    blue_color_ratio = color_ratios["blue"]
    return red_color_ratio, yellow_color_ratio, blue_color_ratio
    
       
def color_ratio(bgr_image, rects=None):
    """
    specified_color_ratio of the image/panel
    :param img: bgr_image
    :param panels: rects of panels
    :return: color_ratio_list
    """
    # 转换颜色空间从BGR到HSV
    hsv_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2HSV)
    
    specified_color = ["red", "yellow", "blue"]
    
    if rects is None:
        red_color_ratio, yellow_color_ratio, blue_color_ratio = compute_color_ratio(hsv_image)
        return red_color_ratio, yellow_color_ratio, blue_color_ratio
    else:
        red_ratios = []
        yellow_ratios = []
        blue_ratios = []
        for i in range(len(rects)):
            x, y, w, h = rects[i]
            hsv_panel = hsv_image[y:y+h, x:x+w]
            red_color_ratio, yellow_color_ratio, blue_color_ratio = compute_color_ratio(hsv_panel)
            red_ratios.append(red_color_ratio)
            yellow_ratios.append(yellow_color_ratio)
            blue_ratios.append(blue_color_ratio)
        return red_ratios, yellow_ratios, blue_ratios 
    

def compute_grayscale_ratio(roi_image, lower_bound, upper_bound):
    """
    给定一个灰度值范围，计算范围内的像素点占比
    :param roi_image: 仅包含ROI区域的bgr图像
    :return background_ratio: 像素点占比
    """
    
    # 转换为灰度图像
    gray = cv2.cvtColor(roi_image, cv2.COLOR_BGR2GRAY)
    
    total_pixels = gray.size # roi_image的总像素
    
    thresholded_image = cv2.inRange(gray, lower_bound, upper_bound)

    # 统计符合条件的像素数量
    thresholded_pixels = np.sum(thresholded_image == 255)
    
    return round(thresholded_pixels/total_pixels, 2)
```

# 强烈色块对比识别

```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
from collections import Counter

def extract_dominant_color(image, k=3):
    """
    提取图像的主色调(背景色)
    
    参数:
        image: 输入图像(BGR格式)
        k: 要提取的色调数量
        
    返回:
        主要颜色的HSV值列表(按出现频率排序)
    """
    # 转换为HSV
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    
    # 重塑为像素点列表
    pixels = hsv.reshape((-1, 3))
    
    # 使用K-means聚类找出主要颜色
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(pixels)
    
    # 获取聚类中心和频率
    counts = Counter(kmeans.labels_)
    total = sum(counts.values())
    
    # 按频率排序
    sorted_colors = sorted([(count/total, center) for center, count in zip(kmeans.cluster_centers_, counts.values())], 
                          key=lambda x: -x[0])
    
    return [color for (freq, color) in sorted_colors]
```

```python
# 提取背景色
dominant_colors = extract_dominant_color(img)
background_hsv = dominant_colors[0]  # 取最主要的颜色作为背景
```

```python
# 创建一个空白的图像，大小为50x50像素，初始化为黑色
image = np.ones((50, 50, 3), dtype=np.uint8) * 255
 
# 将HSV颜色值转换为BGR
bgr_value = cv2.cvtColor(np.uint8([[background_hsv]]), cv2.COLOR_HSV2BGR)[0][0]
 
# 在图像上绘制方形，颜色为转换后的BGR值
cv2.rectangle(image, (0, 0), (49, 49), bgr_value.tolist(), -1)  # (-1)表示填充方形
 
# 显示图像

img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
 
# 使用matplotlib显示图像
plt.imshow(img_rgb)
plt.axis('off')  # 不显示坐标轴
plt.show()
```

```python
def calculate_hsv_contrast(color1, color2):
    """
    计算两个HSV颜色之间的对比度
    
    参数:
        color1, color2: HSV颜色值
        
    返回:
        对比度得分(值越大对比越强烈)
    """
    h1, s1, v1 = color1
    h2, s2, v2 = color2
    
    # 色调差异(考虑色环的循环特性)
    h_diff = min(abs(h1 - h2), 360 - abs(h1 - h2)) / 180.0
    
    # 饱和度差异
    s_diff = abs(s1 - s2) / 255.0
    
    # 明度差异
    v_diff = abs(v1 - v2) / 255.0
    
    # 综合对比度(可调整权重)
    contrast = 0.5 * h_diff + 0.3 * s_diff + 0.2 * v_diff
    
    return contrast
```

```python
def hsv_contrast(hsv_list, background_hsv):
    """
    计算panel HSV均值色与背景HSV色之间的对比度
    
    参数:
        hsv_list: 对比颜色的hsv值列表
        background_hsv: 背景颜色HSV颜色值
        
    返回:
        对比度得分(值越大对比越强烈)
    """
    
    contrast_list = [calculate_hsv_contrast(background_hsv, hsv_mean) for hsv_mean in hsv_list]
    round_contrast_list = np.round(contrast_list, 4)
    
    return round_contrast_list
```
